```python
from nsga_sort import nsga_sort
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

objVals = np.random.random((20,2)) # Raw Values

rank, fronts = nsga_sort(objVals, returnFronts=True) # Fronts used for visualization

# Raw Objective Values
plt.plot(objVals[:,0], objVals[:,1], 'ro')
plt.title('Objective Values of Population')
plt.xlabel('Objective 1')
plt.ylabel('Objective 2')
plt.show()

# Objective Values Sorted by Front
nFronts = len(fronts)
color=iter(cm.rainbow(np.linspace(0,1,nFronts)))
for i in range(nFronts):
  c=next(color)
  sortI = np.argsort((objVals[fronts[i],0])) # To visualize clean fronts
  plt.plot(objVals[fronts[i],0][sortI], objVals[fronts[i],1][sortI], c=c,marker='o',linestyle=':')
plt.title('Population Ranked by NSGA-II')
plt.xlabel('Objective 1')
plt.ylabel('Objective 2')

# Objective Values as Ranked by NSGA-II
for i in range(len(rank)):
    plt.annotate(rank, (objVals[i,0], objVals[i,1]))    

    ''''''
rankedPop = objVals[rank,:]
for i, txt in enumerate(range(len(rank))):
    plt.annotate(txt, (rankedPop[i,0], rankedPop[i,1]))    
plt.show()

```
